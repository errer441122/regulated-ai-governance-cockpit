# Workflow Readiness EDA

Exploratory notebook for the technical lab. The goal is to show how a junior analyst would inspect a simulated AI/data workflow dataset before recommending a controlled pilot.

```python
import pandas as pd

df = pd.read_csv('../data/workflow_signals.csv')
df.head()
```

```python
score_columns = [
    'metadata_quality',
    'process_maturity',
    'stakeholder_clarity',
    'automation_potential',
    'ai_governance_readiness',
]
df[score_columns].describe()
```

```python
weights = {
    'metadata_quality': 0.25,
    'process_maturity': 0.15,
    'stakeholder_clarity': 0.20,
    'automation_potential': 0.15,
    'ai_governance_readiness': 0.25,
}
df['readiness_score'] = sum(df[column] * weight for column, weight in weights.items())
df.sort_values('readiness_score', ascending=False)[['organization', 'sector', 'use_case', 'readiness_score']]
```

## Reviewer Notes

- High readiness does not mean automated approval.
- High sensitivity workflows require stronger human review and privacy checks.
- Low metadata quality should block agentic summarization until source ownership and field definitions are clarified.
